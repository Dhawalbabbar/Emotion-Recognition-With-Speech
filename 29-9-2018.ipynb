```python
from __future__ import print_function
import numpy as np

# matplotlib for displaying the output
import matplotlib.pyplot as plt
import matplotlib.style as ms
ms.use('seaborn-muted')
%matplotlib inline

# and IPython.display for audio output
import IPython.display

# Librosa for audio
import librosa
# And the display module for visualization
import librosa.display
```

```python
sessions=os.listdir('../IEMOCAP_full_release_withoutVideos/IEMOCAP_full_release')
sessions=sessions[3:]
df = pd.DataFrame(columns=['name','det','samprate'])
featu=[]
for _ in sessions:
    direc=os.listdir('../IEMOCAP_full_release_withoutVideos/IEMOCAP_full_release/'+_+'/sentences/wav/')
    for __ in direc:
        files=os.listdir('../IEMOCAP_full_release_withoutVideos/IEMOCAP_full_release/'+_+'/sentences/wav/'+__)
        for i in files:
            print(i)
            y, sr = librosa.load('../IEMOCAP_full_release_withoutVideos/IEMOCAP_full_release/'+_+'/sentences/wav/'+__+'/'+i)
            print(y)
            df=df.append({'name':i,'det':y,'samprate':sr},ignore_index=True)
print(files)
```

```python
df.to_csv('features.csv', sep=',')
```

```python
S = librosa.feature.melspectrogram(df['det'][0], sr=df['samprate'][0], n_mels=128)

# Convert to log scale (dB). We'll use the peak power (max) as reference.
log_S = librosa.power_to_db(S, ref=np.max)

# Make a new figure
plt.figure(figsize=(12,4))

# Display the spectrogram on a mel scale
# sample rate and hop length parameters are used to render the time axis
librosa.display.specshow(log_S, sr=sr, x_axis='time', y_axis='mel')

# Put a descriptive title on the plot
plt.title('mel power spectrogram')

# draw a color bar
plt.colorbar(format='%+02.0f dB')

# Make the figure layout compact
plt.tight_layout()
```
